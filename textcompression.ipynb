```python
import re
from collections import Counter

def compression_ratio(og_text, compressed_text):
    return (len(og_text.encode("ascii")) / len(compressed_text.encode("ascii")));
```

Beginning import statements and compression ratio function

```python
def decode(char_pairs, text):
    for i in range(0, 3):
        for pairs in char_pairs:
            pattern = re.escape(pairs[1]);
            value = pairs[0];
            text = re.sub(pattern, value, text);
    return text;
```

decoding function which uses the regex sub function to replace the
relevant character instances in the given text with the corresponding value in the
2D list used to store the byte-reduction pairs. Also used re.escape so that the
pattern it would replace would be in a usable format. Interestingly, I found that
I could only get this to work if I ran the core part of this function three times,
which is why it is contained in a for loop.

```python
def file_parser(filename):
    file = open(filename, "r");
    next(file)
    text = file.readline();
    next(file);
    char_pairs = [];
    for line in file:
        linepairs = line.split(":=");
        linepairs[1] = linepairs[1].replace("\n", "");
        char_pairs.append(linepairs);
    file.close();
    return (char_pairs, text);
```

Simply opens the file, skips the first line, reads in the text into a string, skips the next line,
and then line by line associates the pairs into a 2D list. Both the text and
mapping are returned. Used the split function with the ":=" as a delimiter, and
removed the \n character from the 2nd value.

```python
text_and_pairs = file_parser("compressed_text.txt");
compressed_text = text_and_pairs[1];
pairs = text_and_pairs[0];
original_text = decode(pairs, compressed_text);

print(compressed_text);
print(original_text);
print(compression_ratio(original_text, compressed_text));
```

# *Part 2!*

```python
def compressor(text):
    #list of all possible ascii characters in a set
    ascii_chars = list(" !\"#$%&'()*+,-./0123456789:;<=>?@ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_`abcdefghijklmnopqrstuvwxyz{|}~");
    #version of the text in a set
    text_set = set(text);
    #set in which to hold the unused letter
    unusedLetters = [];
    #finding all of the unused letters
    for char in ascii_chars:
        if char not in text_set:
            unusedLetters.append(char);

    #empty set to contain all character pairs in the text
    chars = [];
    #loop to make the pairs
    for i in range(len(text)-1):
        pair = text[i] + text[i+1];
        chars.append(pair);

    #here we make a function that will be used later in the function which writes the encryption to a file
    pairs_list = Counter(chars).most_common(len(unusedLetters));
    #uses the Counter class to count the frequency of each pair
    instances = Counter(chars);
    #goes through the unused letter set and replaces the corresponding pair in the text with it
    for i in range(len(unusedLetters)):
        text = text.replace(pairs_list[i][0], unusedLetters[i]);

    #returns all three useful pieces of information
    return [text, pairs_list, unusedLetters];

```

_The logic of this function is as follows:_
* first, it creates a set of all possible ascii characters per Dr. Nguyen's specifications.
* then, it iterates over a set of the provided text and makes a list out of all the characters
  which aren't in the original text
* then, it goes through the text and makes pairs out of every two characters.
* the Counter class then counts the most popular pairs
* Finally, we iterate through the list of unused characters, replacing the most popular pairs
  with the correspondingly indexed unused characters. Then the encrypted text is returned :)

```python
#I chose to use a canto from Dante's Inferno as I've always been a fan of this work.
canto = open("cantoxiii.txt", "r", encoding="ascii");
canto_text = canto.read();
print(canto_text);

#only needs the first part of what compressor returns
compressed_text = compressor(canto_text)[0];
print(compressed_text);
print(compression_ratio(canto_text, compressed_text));
canto.close();
```

# *Part 3!*

```python
def compressedTextWriter(text, file):
    #variable to store relevant info
    compressed_text, pair_list, unusedLetters = compressor(text);

    #begin wri
    # ting to new file
    file.write("[TEXT]\n");
    file.write(compressed_text);
    file.write("[MAPPING]\n");
    #iterate through our list of common character pair and corresponding encoding and write them to file
    for i in range(len(unusedLetters)):
        file.write(str(pair_list[i][0]) + ":=" + unusedLetters[i] + "\n");
```

This file uses the compressor function to get the compressed text, Counter list of all character pair instances, and list of unused letters to write to a file whose filename is given by the user. It mimics the style of the encoded text from the example given to us.

```python
print(canto_text);
compressed_text = open("test_file.txt", "w");
compressedTextWriter(canto_text, compressed_text);
compressed_text.close();

text_and_pairs = file_parser("test_file.txt");
compressed_text = text_and_pairs[1];
pairs = text_and_pairs[0];
original_text = decode(pairs, compressed_text);

print(compressed_text);
print(original_text);
print(compression_ratio(original_text, compressed_text));
```
